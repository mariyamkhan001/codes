```python
#importing nesseacry libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading data set 
df = pd.read_csv('Titanic Datasets.csv')
```

```python
#first few rows of dataset 
print(df.head())
```

```python
# data set information
print(df.info())
```

```python
#print summary of dataset 
# for only numerical data
print(df.describe())
```

```python
print(df.corr())
```

```python
print(df.describe(include='all'))
```

```python
print(df.isnull().sum())
```

```python
# Check for duplicates
print(df.duplicated().sum())
```

```python
# Fill missing values for Age with the mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Fill missing values for Embarked with the most frequent value
#we are Using  the most frequent value (mode). its a simple and unbiased approach compared to other imputation methods, as it reflects the most common category within the data.
#i prefer fillna rather than droping coz dropping will retail valuable data 
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
print(df['Age'])
```

```python
print (df['Embarked'])
```

```python

# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
## We should choose the 'Sex' and 'Embarked' columns for analysis because they provide valuable insights:
# 'Sex' column helps us in understanding gender distribution, imbalance, and survival analysis.
# 'Embarked' column reveals embarkation distribution, geographical insights, and data quality.
```

```python
# Summary Statistics After Cleaning
print("\nSummary statistics after cleaning:")
print(df.describe(include='all'))
```

```python
# Histogram of Age
df['Age'].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Chart of Passenger Counts by Embarked
df['Embarked'].value_counts().plot(kind='bar')
plt.title('Number of Passengers by Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Number of Passengers')
plt.show()
```

```python



# Box Plot of Fare by Pclass
df.boxplot(column='Fare', by='Pclass')
plt.title('Fare Distribution by Class')
plt.suptitle('')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatter Plot of Age vs Fare
df.plot(kind='scatter', x='Age', y='Fare')
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Correlation Analysis
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading 
df = pd.read_csv('Titanic Datasets.csv')
```

```python
#  Bar Chart
#Heren we Visualize the number of passengers who embarked from each port (C, Q, S).
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=df, palette='viridis')
plt.title('Number of Passengers by Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#  Scatter Plot: Age vs Fare
#Explore the relationship between passengers age and the fare they paid, with the color representing passenger class reson is that we wannna observe how fare vareis with age and class
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=df, hue='Pclass', palette='deep')
plt.title('Age vs Fare with Passenger Class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Scatter Plot
#We wanna see here how survival rates varied by age and gender.
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Survived', data=df, hue='Sex', palette='coolwarm')
plt.title('Age vs Survival by Gender')
plt.xlabel('Age')
plt.ylabel('Survival')
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=df, palette='coolwarm')
plt.title('Number of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#  Bar Chart
#Here we wanna understand how passenger class affected survival chances

plt.figure(figsize=(8, 6))
sns.barplot(y='Pclass', x='Survived', data=df, palette='Set2', orient='h')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Survival Rate')
plt.ylabel('Passenger Class')
plt.show()
```
